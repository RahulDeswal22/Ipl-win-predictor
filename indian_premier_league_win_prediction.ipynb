```python
import pandas as pd
import numpy as np
```

loading the datasets and read the data from csv file using pandas library

```python
match = pd.read_csv('matches.csv')
delivery = pd.read_csv('deliveries.csv')
```

```python
total_score_df = delivery.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
```

```python
total_score_df = total_score_df[total_score_df['inning'] == 1]
```

```python
total_score_df
```

adding two new columns match_id and total_run to the data frame

```python
match_df = match.merge(total_score_df[['match_id','total_runs']], left_on='id', right_on='match_id')
```

```python
match_df
```

```python
#get the unique name of each team
match_df['team1'].unique()
```

```python
#selecting the updated teams because same teams have changed their name 

teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
    
]
```

```python
#replacing the old name of team with the new name

match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
#updating the match data frame
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
#checking the new shape
match_df.shape
```

```python
match_df
```

```python
match_df = match_df[match_df['dl_applied'] == 0]
```

```python
#addding some more columns 
match_df = match_df[['match_id','city','winner','total_runs']]
```

```python
#merging the new match_df dataframe to the delivery_df
delivery_df = match_df.merge(delivery,on='match_id')
```

```python
delivery_df = delivery_df[delivery_df['inning'] == 2]
```

```python
delivery_df.shape
```

```python
delivery_df
```

```python
#add a new column cuurent score and current score will be calculated as
delivery_df['current_score'] = delivery_df.groupby('match_id').cumsum()['total_runs_y']
```

```python
delivery_df
```

```python
#add a new column runs left and runs left will be calculated as
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
delivery_df
```

```python
##add a new column balls left and Balls left will be calculated as 

delivery_df['balls_left'] = 126 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df
```

```python
#fill nan missing values
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
wickets = delivery_df.groupby('match_id').cumsum()['player_dismissed'].values

#adding a new column wickets

delivery_df['wickets'] = 10 - wickets
delivery_df.head()
```

```python
#checking the tail values of updated delivery_df dataset
delivery_df.tail()
```

```python
# crr = runs/overs

delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
#for value of rrr we do

delivery_df['rrr'] = (delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
delivery_df
```

```python
#result of a team 
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
#adding column result to the delivery_df 
delivery_df['result'] = delivery_df.apply(result, axis =1)
```

```python
delivery_df
```

```python
#Extracting only useful features and Our final_df will be

final_df = delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
```

```python
final_df
```

```python
#Shuffle the final data 
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.sample()
```

```python
final_df.isnull().sum()
```

```python
#drop null values
final_df.dropna(inplace=True)
```

```python
final_df = final_df[final_df['balls_left'] !=0]
```

```python
#Applying train test split
X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state=1) 
```

```python
X_train
```

```python
#We import ColumnTransformer and One Hot Encoding
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False, drop='first'),['batting_team','bowling_team','city'])
]
,remainder = 'passthrough')
```

```python
#We import LogisticRegression, Pipeline and fit our training data in pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train, y_train)
```

```python
#calculating the y_pred 
y_pred = pipe.predict(X_test)
```

```python
#we import accuracy score from scickit learn
from sklearn.metrics import accuracy_score

#Now we check accuracy score
accuracy_score(y_test,y_pred)
```

```python
pipe.predict_proba(X_test)[10]
```

```python
def match_summary(row):
    print("Batting Team-" + row['batting_team'] + " | Bowling Team-" + row['bowling_team'] + " | Target- " + str(row['total_runs_x']))
```

```python
def match_progression(x_df,match_id,pipe):
    match = x_df[x_df['match_id'] == match_id]
    match = match[(match['ball'] == 6)]
    temp_df = match[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr']].dropna()
    temp_df = temp_df[temp_df['balls_left'] != 0]
    result = pipe.predict_proba(temp_df)
    temp_df['lose'] = np.round(result.T[0]*100,1)
    temp_df['win'] = np.round(result.T[1]*100,1)
    temp_df['end_of_over'] = range(1,temp_df.shape[0]+1)
    
    target = temp_df['total_runs_x'].values[0]
    runs = list(temp_df['runs_left'].values)
    new_runs = runs[:]
    runs.insert(0,target)
    temp_df['runs_after_over'] = np.array(runs)[:-1] - np.array(new_runs)
    wickets = list(temp_df['wickets'].values)
    new_wickets = wickets[:]
    new_wickets.insert(0,10)
    wickets.append(0)
    w = np.array(wickets)
    nw = np.array(new_wickets)
    temp_df['wickets_in_over'] = (nw - w)[0:temp_df.shape[0]]
    
    print("Target-",target)
    temp_df = temp_df[['end_of_over','runs_after_over','wickets_in_over','lose','win']]
    return temp_df,target
```

```python
#Using match_progression and match_summary function we can call a specific match id resulting in the temp_df
temp_df,target = match_progression(delivery_df,74,pipe)
temp_df
```

```python
#using the “matplotlib” library to create a combined plot that visualizes various aspects of the cricket match progression
import matplotlib.pyplot as plt
plt.figure(figsize=(18,8))
plt.plot(temp_df['end_of_over'],temp_df['wickets_in_over'],color='yellow',linewidth=3)
plt.plot(temp_df['end_of_over'],temp_df['win'],color='#00a65a',linewidth=4)
plt.plot(temp_df['end_of_over'],temp_df['lose'],color='red',linewidth=4)
plt.bar(temp_df['end_of_over'],temp_df['runs_after_over'])
plt.title('Target-' + str(target))
```
